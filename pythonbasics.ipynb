# Python Basics Notebook

This notebook introduces 9th and 10th graders to the basics of Python.

```
# 1. Introduction to Python
print("Hello, world! My name is Riya!")  # This prints a message to the screen
```

```
# 2. Variables & Data Types
name = "Alice"  # String
age = 14  # Integer
height = 5.4  # Float
is_student = True  # Boolean

# Displaying variable values
print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Is student?", is_student)
```

```
# 3. Basic Operations
sum_result = 5 + 3  # Addition
product = 4 * 2  # Multiplication
message = "Python is " + "fun!"  # String concatenation
print(sum_result, product, message)
```

```
# 4. Control Flow
x = 10
if x > 5:
    print("x is greater than 5")
else:
    print("x is 5 or less")
```

```
# 5. Loops
for i in range(5):
    print("Loop iteration:", i)
```

```
# 6. Functions
def greet(name):
    return "Hello, " + name + "!"

print(greet("Alice"))
def Operations (a, b):
    sum = a + b
    product = a * b
    if product>sum: 
        print("product is greater than sum")
    else:
        print("sum is greater than product")
    return sum, product

print(Operations(5, 3))

def Pythag(a, b):
    c = (a**2 + b**2)**0.5
    return c
print(Pythag(3, 4))
```

```
# 7. Lists & Dictionaries
fruits = ["apple", "banana", "cherry"]
print("First fruit:", fruits[0])

student = {"name": "Alice", "age": 14}
print("Student Name:", student["name"])
```
